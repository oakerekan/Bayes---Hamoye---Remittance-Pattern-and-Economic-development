## Data Cleaning - Remittance Outflows

```python
import numpy as np 
import pandas as pd 
import warnings
```

```python
# Filepath
file_path = '/content/outward_remittance_flows.xlsx'
outflow = pd.read_excel(file_path,index_col=0)
outflow
```

```python
# Rename some outflow columns
outflow = outflow.rename(columns={'% of GDP in 2021': '%GDP_2021'})
# Selecting country names only
outflow = outflow.iloc[:214, :-2]
outflow
```

```python
# Display the dataset information
outflow.info()
```

```python
# replace empty rows with NaN values
outflow.replace('', np.nan, inplace=True)
```

```python
# Check for missing values
outflow.isnull().sum()
```

```python
# Remove all rows that have missing values
outflow.dropna(inplace=True)
outflow.head()
```

```python
outflow.shape
```

```python
outflow.dtypes
```

```python
# # Convert string object to float64
outflow = outflow.apply(pd.to_numeric, errors='coerce')
outflow.head(10)
```

```python
outflow.dtypes
```

```python
# Save cleaned remittance outflow data to CSV file
outflow.to_csv('remittance_outflows_clean.csv', index=True)
```
